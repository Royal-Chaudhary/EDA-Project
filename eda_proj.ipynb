```python
def energyCat(energy):
  if energy<=0.35:
    return "dull"
  elif energy>0.35 and energy<0.75:
    return "moderate"
  else:
    return "vibrant"
dfx['energy_category'] = dfx['energy'].apply(energyCat)
```

```python
dfx_copy = dfx.copy()
```

```python
dfx_copy_1 = dfx.copy()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)

dfx_copy.loc[(dfx_copy['explicit']==True),'T'] =1
dfx_copy.loc[(dfx_copy['explicit']==False),'F'] =1

markers = ['True', 'False']
sizes = [dfx_copy['T'].value_counts()[1],
         dfx_copy['F'].value_counts()[1],
         ]
plt.title("The frequency of Explicit")
plt.pie(sizes, labels=markers,autopct='%1.1f%%')

plt.subplot(1,2,2)
sns.set_style('ticks')
sns.countplot(data=dfx_copy,x='explicit')

plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)

dfx_copy_1.loc[(dfx_copy_1['energy_category']=='dull'),'D'] =1
dfx_copy_1.loc[(dfx_copy_1['energy_category']=='moderate'),'M'] =1
dfx_copy_1.loc[(dfx_copy_1['energy_category']=='vibrant'),'V'] =1
markers = ['Dull', 'Moderate','Vibrant']
sizes = [dfx_copy_1['D'].value_counts()[1],
         dfx_copy_1['M'].value_counts()[1],
        dfx_copy_1['V'].value_counts()[1],
         ]
plt.title("The frequency of energy_category")
plt.pie(sizes, labels=markers,autopct='%1.1f%%')

plt.subplot(1,2,2)
sns.set_style('ticks')
sns.countplot(data=dfx_copy_1,x='energy_category')

plt.show()
```
